```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data =pd.read_csv(r'C:\Users\HP\Desktop\datasets%2F596958%2F1073629%2FPlacement_Data_Full_Class.csv')
data
```

```python
data.describe()
```

```python
y= data['mba_p']
x1=data[['ssc_p','hsc_p','degree_p']]
```

```python
x= sm.add_constant(x1)
results= sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['ssc_p'],y)
yhat_ssc_p=37.6529+0.0965*data['ssc_p']
yhat_hsc_p=37.6529+0.0860*data['hsc_p']
yhat_degree_p=37.6529+0.1872*data['degree_p']
fig = plt.plot(data['degree_p'],yhat_degree_p,lw=2,c='yellow')
fig = plt.plot(data['ssc_p'],yhat_ssc_p,lw=2,c='red')
fig = plt.plot(data['hsc_p'],yhat_hsc_p,lw=2,c='green')
plt.xlabel('ssc_p',fontsize=20)
plt.ylabel('mba_p',fontsize=20)
plt.show()
```
